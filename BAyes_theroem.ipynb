```python
d = pd.DataFrame({
    "CGPA":['g9','g8','g9','l8','g8','g9','l8','g9','g8','g8'],
    "inter":['y','n','n','n','y','y','y','n','y','y'],
    "pk":['+++','+','==','==','+','+','+','+++','+','=='],
    "cs":['g','m','p','g','m','m','p','g','g','g'],
    "job":['y','y','n','n','y','y','n','y','y','y']
})
```

```python
d
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
from sklearn.metrics import classification_report,confusion_matrix
```

```python
cgpa_l= LabelEncoder()
inter_l= LabelEncoder()
pk_l= LabelEncoder()
cs_l= LabelEncoder()
job_l= LabelEncoder()
cgpa_l.fit(d['CGPA'])
inter_l.fit(d['inter'])
pk_l.fit(d['pk'])
cs_l.fit(d['cs'])
job_l.fit(d['job'])
```

```python
d['inter']=inter_l.transform(d['inter'])
d['pk']=pk_l.transform(d['pk'])
d['cs']=cs_l.transform(d['cs'])
d['job']=job_l.transform(d['job'])
d['CGPA']=cgpa_l.transform(d['CGPA'])
```

```python
d
```

```python
model=GaussianNB()
x_train,x_test,y_train,y_test=train_test_split(d[['CGPA','inter','pk','cs']],d['job'],test_size=0.2)
model.fit(x_train,y_train)
```

```python
print("x_train\n",x_train)
print("y_trsin\n",y_train)
print("x_test\n",x_test)
print("y_test\n",y_test)
```

```python
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
