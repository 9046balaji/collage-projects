## Structure the results

### Subtask:
Organize the extracted association rules into a structured format (e.g., a pandas DataFrame) including the antecedent, consequent, support, confidence, and lift for each rule.


**Reasoning**:
I will create a list to store the structured rules, iterate through the association results, extract the required information (antecedent, consequent, support, confidence, lift) for each rule, handle the frozensets by converting them to strings, and append the structured data to the list. Finally, I will convert this list into a pandas DataFrame.



```python
from apyori import apriori

records = []
for i in range(1, 7501):
    records.append([str(store_data.values[i, j]) for j in range(0, 20)])

association_rules = apriori(records, min_support=0.001, min_confidence=0.05, min_lift=1.5, min_length=2)
association_results = list(association_rules)

structured_rules = []
for item in association_results:
    if len(item[2]) > 0:
        for relation in item[2]:
            antecedent = ', '.join(list(relation[0]))
            consequent = ', '.join(list(relation[1]))

            support = item[1]
            confidence = relation[2]
            lift = relation[3]

            structured_rules.append({
                'Antecedent': antecedent,
                'Consequent': consequent,
                'Support': support,
                'Confidence': confidence,
                'Lift': lift
            })

rules_df = pd.DataFrame(structured_rules)
display(rules_df.head())
```
